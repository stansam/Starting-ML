```python
from sklearn import svm
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pickle
import warnings
# import streamlit as st
# from pyngrok import ngrok

warnings.filterwarnings("ignore")
dataset = pd.read_csv('/content/diseases_training.csv')
```

```python
dataset.dropna()
```

Data preprocessing

```python
dataset["diagnostics"] = dataset['diagnosis'].round().astype(int)
```

Dataframe snapshot

```python
dataset
print(dataset.columns)
```

```python
selected_symptoms = ['itching', 'skin_rash', 'continuous_sneezing', "chills", 'joint_pain', 'vomiting', 'fatigue', 'weight_loss', 'cough', 'high_fever', 'headache', 'yellowish_skin', 'prognosis', 'diagnosis' ]
new_dataset = dataset[selected_symptoms]
new_dataset
```

```python
new_dataset.describe()
```

```python
new_dataset['prognosis'].unique()
```

```python
class_count = new_dataset['prognosis'].value_counts()
class_count
```

```python
y = new_dataset['prognosis']
X = new_dataset.drop(['diagnosis', 'prognosis'], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=2 )
```

```python
model = svm.SVC(kernel='linear')
```

```python
model.fit(X_train, y_train)
```

```python
prediction = model.predict(X_test)
prediction
```

```python
test_accuracy = accuracy_score(y_test, prediction)
test_accuracy
```

```python
train_pred = model.predict(X_train)
train_pred
train_accur = accuracy_score(y_train, train_pred)
train_accur
```
